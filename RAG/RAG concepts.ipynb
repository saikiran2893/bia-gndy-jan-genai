What is RAG:

1. RAG = Retrival + Generation. We don't rely om LLM for all the information or facts. we fetch the information from the knowledge source and feed that information(Context) to the model and get the final data.

2. Why it matters?

- LLM have cutoff date, they can hallucinate
- LLM doesn't have knowledge on the domain or business data that get updated frequently
- LLM can't able to answer from various retirved data

3. Usecase:
 - Chatbots(Customer Support, Q&A over documents)
 - Policy related information Retrival
 - Traning Document / User guide Q&A

Architecture:

User query -> Embeded -> Vector Search on DB -> Top -k Chunks -> [Context + query] -> LLM -> Answer

1. Embedding:  https://developers.openai.com/api/docs/guides/embeddings/#embedding-models.  https://huggingface.co/spaces/mteb/leaderboard

- Dense vectors that represent words, sentences, or documents in a continuous vector space.
Example: Cat and Dog will have closer vectors compared to Cat and Car.

- We have embeddings model ( Open AI(paid), google(paid and opensource), bedrock(paid and opensource), huggingface (open source), sentence transformer(open-source))

- Dimension (384, 768, 1024, 1536) -> 384 (miniLM) , 768 (majority), 1536(open AI)



2. Similarity Metrics:

- 1. Cosine Similarity: It measures the angle between two vectors(Query vector and KB vectors) -> Cos(90)-> 0 -> cos(0) -> 1 . When you more care about meaning than the length

- 2. Dot Product Fast, When vectors are normalised , when you care about meaning and length (Dot product)

- 3. Euclidean Distance: numbers

3. Vector Databases:

 - The tradational DB which uses only exact key word search , sematic search is not possible
 - Vector DB supports the semantic search:
    - we store the embeddings and the document chunks(KB)

- It can support:
  - Vector Search( cosine /IP)
  - Indexing (ANN, Flat Index, HNSW, IVF-Flat)
  - Scalable and persistance
  - Metadata filtering

- FAISS (Vector Dadatabase) , Pinecone, Weaviate, Chroma, Milvus



## Text - RAG Pipeline

1. Core Stages:

 - We have Knowledge base documents
 - Chunks the each document from the knowledge base
 - Index chunks with embeddings
 
 - Retrive the relevant chunks
 - Generate the response with LLM with query and retived chunks

Chunking Stratergies:

1. Fixed - length Chunking:  where you split the document based on the character or token count . This simple but it cuts the sentences or ideas into half.
2. Sentence / semantic Chunking : Chunks are aligned to the sentences or paragraph . Better Semantics and logic
3. Recursive Chunking : Split by Strucutre( heading, bullets, paragraph) , to do it we need better parsers and it hold good semantic coherence.
4. Sliding window : Fixed Size + Overlap : Balances context and simplicity ..


Best Practice : Combine Recursive + sliding window(overlap) for better context and boundaries

# Multi modal RAG

Multi modal RAG is an AI architecture where a system can store, retrive  and understand multiple types of data and with help of multi-modal LLM it ables generate answers.

Modalities

- Text - PDF, documents and articles
- Images - photos, diagrams, charts
- Audio - recordings
- Videos - youtube, surveillance
- Tables - strucutred data

Tradational RAG system where it works with the text , tables

Multimodal RAG you expand this to include, images, audio and videos

Embeddings:

-  Dog or cat -> name of the animal, they mentioned what it was doing in that image( this helps to learn the description of the image)
