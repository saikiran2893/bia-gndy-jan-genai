```python
# import ssl
# ssl._create_default_https_context = ssl._create_unverified_context


## https://pypi.org/project/detoxify/


from detoxify import Detoxify

detect = Detoxify('original')
```

```python
customer_responses = [
    ("Safe & helpful",
     "I completely understand your frustration. Let me look into your account and find the best solution for you."),
    
    ("Dismissive",
     "Your issue is not our problem. Read the terms and conditions you agreed to."),
    
    ("Passive aggressive",
     "As I have already explained multiple times, the refund policy clearly states no exceptions."),
    
    ("Direct insult",
     "This is a ridiculous complaint. You're being completely unreasonable and wasting our time."),
    
    ("Escalation language",
     "If you keep harassing our staff with these demands, we will be forced to ban your account."),
    
    ("Professional boundary",
     "I'm not able to assist with that request. Is there something else I can help you with today?"),
]

TOXICITY_THRESHOLD = 0.3


for label, response in customer_responses:
    scores = detect.predict(response)
    toxicity_score = scores['toxicity']
    print(f"Response: '{response}'\nToxicity Score: {toxicity_score:.4f} - Label: {label}")
    if toxicity_score > TOXICITY_THRESHOLD:
        print("-> This response is flagged as potentially non-compliant.\n")
    else:
        print("-> This response is considered compliant.\n")

```
