```python
from typing_extensions import Optional
from pydantic import BaseModel
```

```python
class User(BaseModel):
    id: int
    name: str
    email: str
    age: int
    city: Optional[str] = None
```

```python
User(id=1, name="Alice", email="alice@example.com", age=30, city="New York")
```

```python
User(id=1, name="Alice", email="alice@example.com", age=30)
```

```python
User(id=1, name="Alice", email="alice@example.com")
```

- Optional[type] 
- Default value ( = None or = True)
-  Requried field are mandatory
- Pydantic will validate types and values that are provided

```python
User(id="a", name="Alice", email="alice@example.com", age="30")
```

```python
from typing import List

class Classroom(BaseModel):
    room_number: int
    students: List[str]
    capcity: int

Classroom(room_number=101, students=["Alice", "Bob","Charlie"], capcity=3)
```

```python
## Nested Models validation



from pydantic import BaseModel

class Address(BaseModel):
    street: str
    city: str
    zip_code: str


class Customer(BaseModel):
    id: int
    name: str
    Address: Address


customer_details = Customer(id="2",name = "Alice",Address={"street":"mainroad","city":"New York","zip_code":"10001"})


##  LLM Call/ Database inserts / API Call / Print:


print(customer_details.id)
print(customer_details.name)
print(customer_details.Address.street)
print(customer_details.Address.city)
print(customer_details.Address.zip_code)
```

```python
## Pydantic Fields

from pydantic import BaseModel, Field

class Product(BaseModel):
    id:int
    Name:str=Field(min_length=3, max_length=50)
    price:float=Field(gt=0)
    quantity:int=Field(ge=0)
    serialnumber:str=Field(default= "00000")


product = Product(id=1, Name="Laptop", price=999.99, quantity=10,serialnumber="12345")
product

```
