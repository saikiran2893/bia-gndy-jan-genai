## Tools

# 1. Concepts of Tools

1. A tool is 
    - A function
    - Plus metadata(name, description, arugmentation schema)
    - This allows LLM to make a call


2. An Agent:
    - User send one request
    - Agent decides which tool to call(support of LLM)
    - Toll call will provide the output which needs to be interpreted by agent
    - And responds it back to the user


why tools needed:

 - LLM are nothing but next work token generators
 - with tools they can:
    - call API
    - seach the web
    - Query database
    - Trigger workflow etc.

```python
from langchain.tools import tool

@tool
def multiply(a:int, b:int)->int:
    '''this tool is for multiplication of two numbers and it takes two integers as input and returns the product of the two numbers as single integer output'''
    return a * b
```

```python
multiply.invoke({"a":10, "b":20})
```

```python
multiply.invoke({"a":3, "b":4})
```

```python
multiply.run({"a":10, "b":20})
```

```python
multiply.name, multiply.description, multiply.args
```

```python
from typing import List
@tool
def add(a:List) -> int:
    '''this tool is for addition of a list of numbers and it takes a list of integers as input and returns the sum of the numbers as single integer output'''
    return sum(a)


## LLM : give me sum of all these numbers 1,2,3,4,5.count
## LLM : to call add tool and pass the list of numbers as input to the tool [1,2,3,4,5]
add.invoke({"a":[1,2,3,4,5]})
```
