## Building the First Agent

```python
def add_numbers(a: float, b: float) -> float:
    """Adds two numbers together.
    Args:
        a (float): The first number.
        b (float): The second number.
    Returns:
        float: The sum of a and b.
    """
    return a + b

def subtract_numbers(a: float, b: float) -> float:
    """Subtracts b from a.
    Args:
        a (float): The first number.
        b (float): The second number.
    Returns:
        float: The result of a - b.
    """
    return a - b

def multiply_numbers(a: float, b: float) -> float:
    """Multiplies two numbers together.
    Args:
        a (float): The first number.
        b (float): The second number.
    Returns:
        float: The product of a and b.
    """
    return a * b

def divide_numbers(a: float, b: float) -> float:
    """Divides a by b.
    Args:
        a (float): The numerator.
        b (float): The denominator.
    Returns:
        float: The result of a / b.
    """
    if b == 0:
        raise ValueError("Cannot divide by zero.")
    return a / b

## Building tools

def get_weather(city:str,unit:str = 'celsius'):
    """
    This is a tool to get the weather of a city. It takes in the city name and the unit of temperature (celsius or fahrenheit) and returns the current weather of the city in the specified unit.
    Args:
    city (str): The name of the city to get the weather for.
    unit (str): The unit of temperature to return the weather in. Can be either 'celsius' or 'fahrenheit'. Default is 'celsius'.
    Returns:   
    str: A string describing the current weather in the specified city and unit.
    """
    # For demonstration purposes, we will return a dummy weather report
    if unit == 'celsius':
        return f"The current weather in {city} is 25 degrees Celsius with clear skies."
    elif unit == 'fahrenheit':
        return f"The current weather in {city} is 77 degrees Fahrenheit with clear skies."
    else:
        return "Invalid unit. Please specify either 'celsius' or 'fahrenheit'."
```

```python
from google.adk.agents import LlmAgent
from google.adk.models.lite_llm import LiteLlm
from google.genai import types

agent = LlmAgent(name = 'calculator_agent',model=LiteLlm(model='gpt-4o-mini'),
                instruction="You are an helpful calculator agent, Always use tools to calcualte the result. Display you work in detail",
                description="This agent is designed to perform calculations and provide detailed explanations of the steps taken to arrive at the final result",
                tools=[add_numbers, subtract_numbers, multiply_numbers, divide_numbers],
                sub_agents=[])     


    

## Run Agents

import asyncio
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types


session_service = InMemorySessionService()
runner = Runner(agent = agent, app_name="app",session_service=session_service)

session = await session_service.create_session(user_id="user123", app_name="app")


async for event in runner.run_async(
    user_id="user123",
    session_id=session.id,
    new_message=types.Content(role="user", parts=[types.Part(text="Can you provide me. sum of 5 and 10?")]),):
        
        print(event)
        print("/n/n *******")

        if event.is_final_response:
            print("Agent response: ", event.message.parts[0].text)




```
