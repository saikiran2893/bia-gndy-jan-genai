## Langgraph


- Langgraph a library or framework based on python used to build stateful and multi-step workflows. It involves:
    - LLM
    - Tools(API, Database, RAG, search, etc)
    - Custom Python logic
    - Memory /State


- In Langchain you basically chain the flow here you build the directed graph

# Langchain vs Langgraph:

- A chain will be linear : Step1 -> Step2 -> step3 -> Done
Example: If user ask to do mathematical calculation then -> call the Calculator Tool -> if it fails it retries or do fallback -> show the result to the user


But in reworld Agentic AI we need branching, looping, state management instead of chain that is been linear

-> if user ask any question: 
-> it should route to the break the question and make sure all the parts of the question been answered using the various tools
-> interact all those results
-> show the result to the user




## What Make Langgraph powerful?

- Build AI Agents
- Create multi-step reasoning
- Combine LLM + tool + logic clearly 
- you execute the flow deterministically, not just a single LLM call
- You add fall back mechanism (retry logic), validations 

## Langgraph Core


# Graph: (Blueprint of your AI system)

- Entire workflow it has :
 - Nodes
 - Edges
 - Entry point
 - Finish point
 - State Schema



 # Node:

 - It is function that get the current state -> apply the function -> Returns the update to the State

 - State are in the dictionary format

 - Call LLM, Call tool, Validate any condition / output


 # Edge:

- It defines where the execution goes next
- They repesent the direction

# State:

- State is a data / messages that get shared across the nodes.

 - it will be in the Dict, TypeDict ,etc (Pydantic model)


 Why we need state:

 - Consider you have 3 functions:
    - Function 1, function 2, Function 3 -> you define Variable and you make you pass the variable each and everytime you call

    - State are shared objects:

        - Intial State -> Node A (State updated) -> Node B (State updated) -> Node C ( State get updated) -> end node

# Entry Point:

- Start node where the execution starts


# Finish Point:

- End node where the execution stops


-> To run the workflow 1. Compile 2. invoke


## Advantages of using Langgraph than Langchain:

1. Transperancy - Each node you see exactly what happend
2. Control - Node can able to decide what to be done like tool call, LLM, Other nodes, human intervention, etc.
3. Resuse - Same node can be reused at multiple times
4. Debugging - you can easily debug the execution i.e output of each node
